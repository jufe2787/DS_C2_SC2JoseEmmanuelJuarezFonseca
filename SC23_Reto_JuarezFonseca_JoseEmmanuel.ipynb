Creacion de 4 data frames a partir de los archivos CSV

```python
from numpy import NaN
import pandas as pd

Test = pd.read_csv("/content/gdrive/MyDrive/Colab Notebooks/Test.csv")
Test2 = pd.read_csv("/content/gdrive/MyDrive/Colab Notebooks/Test2.csv")
Train = pd.read_csv("/content/gdrive/MyDrive/Colab Notebooks/Train.csv")
Train2 = pd.read_csv("/content/gdrive/MyDrive/Colab Notebooks/Train2.csv")
```

Conjuntar informacion y generar data frame unico con las columnas necesarias.

```python
Dframe_employees = pd.concat([Test, Train], ignore_index = True)
Dframe_os = pd.concat([Test2, Train2], ignore_index = True)

result = pd.merge(Dframe_employees, Dframe_os, on = "Employee_ID")

Data_Frame = result[["Employee_ID", "Age", "Education_Level", " Computer_OS", " Mobile_OS", " Computer_tickets"]]
```

Asignacion de "no registrado" a registros sin sistema operativo.

```python
Data_Frame[" Computer_OS"] = Data_Frame[" Computer_OS"].fillna("no registrado")
Data_Frame[" Mobile_OS"] = Data_Frame[" Mobile_OS"].fillna("no registrado")
```

Porcentaje de personas que usan Windows / Mac / Linux / no registrado como su sistema operativo en su computadora.

```python
from numpy.lib.function_base import percentile
Porcentaje_OS = Data_Frame[[" Computer_OS"]].groupby(" Computer_OS").size() / Data_Frame[" Computer_OS"].count() * 100
Porcentaje_OS
```

El porcentaje de personas que usan iOS / Android / No registrado como su sistema operativo en su dispositivo m√≥vil.

```python
Porcentaje_Mobile_OS = Data_Frame[[" Mobile_OS"]].groupby(" Mobile_OS").size() / Data_Frame[" Mobile_OS"].count() * 100
Porcentaje_Mobile_OS
```

La edad promedio de la gente que utiliza cada sistema operativo de escritorio.

```python
Promedio_Edad = Data_Frame[[" Computer_OS", "Age"]].groupby(" Computer_OS").mean()
Promedio_Edad
```

El nivel educativo promedio de la gente que utiliza cada sistema operativo de escritorio

```python
Promedio_Nivel_Ed = Data_Frame[[" Computer_OS", "Education_Level"]].groupby(" Computer_OS").mean()
Promedio_Nivel_Ed
```

El n√∫mero total de tickets generados por la gente con cada sistema operativo de escritorio

```python
Total_Tickets = Data_Frame[[" Computer_OS", " Computer_tickets"]].groupby(" Computer_OS").sum()
Total_Tickets
```

Excel

```python
with pd.ExcelWriter("resultadosReto.xlsx") as writer:
  Data_Frame.to_excel(writer, sheet_name = "Data Frame")
  Porcentaje_OS.to_excel(writer, sheet_name = "Porcentaje OS")
  Porcentaje_Mobile_OS.to_excel(writer, sheet_name = "Porcentaje Mobil OS")
  Promedio_Edad.to_excel(writer, sheet_name = "Promedio Edad")
  Promedio_Nivel_Ed.to_excel(writer, sheet_name = "Promedio Nivel Educativo")
  Total_Tickets.to_excel(writer, sheet_name = "Total Tickets")
```
